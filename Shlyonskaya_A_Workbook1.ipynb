# Задание 4.3.1

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерируем массив случайных значений из интервала (0, 1)
np.random.seed(0)  # Для воспроизводимости результатов
data = np.random.rand(100)  # Генерируем 100 случайных значений

# Рассчитываем среднее значение и медиану
mean_value = np.mean(data)
median_value = np.median(data)

# Выводим результаты
print("Среднее значение:", mean_value)
print("Медиана:", median_value)

# Строим точечную диаграмму рассеяния
plt.scatter(range(len(data)), data, alpha=0.5)
plt.axhline(mean_value, color='red', linestyle='--', label='Среднее значение')
plt.axhline(median_value, color='green', linestyle='--', label='Медиана')
plt.legend()
plt.xlabel('Индекс элемента')
plt.ylabel('Значение')
plt.title('Точечная диаграмма рассеяния')
plt.show()
```

# Задание 4.3.2

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем массив значений x от 1 до 10
x = np.arange(1, 10)

# Определяем функцию
def my_function(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x)**2) / abs(1 - np.sin(x)**3) + np.log(np.abs(2 * x))

# Вычисляем значения функции для всего массива
y = my_function(x)

# Выделяем срез первой половины массива
x_slice = x[:5]
y_slice = y[:5]

# Строим графики
plt.figure(figsize=(12, 5))

# График для всего массива
plt.subplot(121)
plt.plot(x, y, label='Функция')
plt.xlabel('x')
plt.ylabel('Значение функции')
plt.title('График функции для всего массива')
plt.legend()

# График для среза
plt.subplot(122)
plt.scatter(x_slice, y_slice, label='Срез', color='red', marker='o')
plt.xlabel('x')
plt.ylabel('Значение функции')
plt.title('Точечный график для среза')
plt.legend()

plt.tight_layout()
plt.show()
```

# Задание 4.3.3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz

# Создаем массив значений x на интервале (0, 10) с шагом 1
x = np.arange(0, 10)

# Вычисляем значения функции |cos(xe^(cos(x)+ln(x+1)))|
y = np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

# Строим график функции с заливкой площади
plt.plot(x, y, label='|cos(xe^(cos(x)+ln(x+1)))|', color='blue')
plt.fill_between(x, 0, y, alpha=0.2, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции с заливкой площади')
plt.legend()

# Вычисляем площадь под графиком с использованием метода трапеций
area = trapz(y, x)
print("Площадь под графиком:", area)

# Отображаем график
plt.show()
```
